## Part 3: Statistical Testing
### Chi-Square Test of Independence
#### Gender vs. Most Used Dating App

```python
# chi-square test of independence
from scipy.stats import chi2_contingency

# create a contingency table
cont1 = pd.crosstab(df['gender'], df['most_used'])
print('contingency 1:')
print(cont1.to_string())
print('#'*60)

# get row percentages
cont1_rowp = cont1.div(cont1.sum(axis=1), axis=0) * 100
print('\nrow percentages cont1:')
print(cont1_rowp.to_string())
print('#'*60)

# perform chi-square test
chi2, p, dof, expected = chi2_contingency(cont1)

print(f"\nChi-squared value: {chi2}")
print(f"P-value: {p}")
print(f"Degrees of freedom: {dof}")
```

```python
# create a contingency table
cont2 = pd.crosstab(df['race'], df['most_used'])
print('contingency 2:')
print(cont2.to_string())
print('#'*60)

# get row percentages
cont2_rowp = cont2.div(cont2.sum(axis=1), axis=0) * 100
print('\nrow percentages cont2:')
print(cont2_rowp.to_string())
print('#'*60)

# perform chi-square test
chi2, p, dof, expected = chi2_contingency(cont2)

print(f"\nChi-squared value: {chi2}")
print(f"P-value: {p}")
print(f"Degrees of freedom: {dof}")
```
